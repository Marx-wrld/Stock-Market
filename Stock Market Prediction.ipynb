### Setting up our target(Will our price go up or down tomorrow)

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500['Target'] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
###Is tomorrow's price greater than today's price, 0 when price goes down
```

```python
sp500 = sp500.loc["1990-01-01":].copy()
### removing the data before 1990
```

### Training our initial ML model

```python
from sklearn.ensemble import RandomForestClassifier
### Works by training a bunch of individual decision trees with randomized parameters
### And then averaging the results from those decision trees.
### Because of this randomforest is resistant to overfitting, they also run relatively quickly and they can pick up non-linear tendencies in the data
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
### random state helps us in a way if we rerun the model twice we get the same results
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score
### Tests whether when we said the market will go up, did the market really go up
preds = model.predict(test[predictors])
```

```python
preds
## Predictions are in a numpy array which is a little bit hard to work with, so we're actually goin to urn this into a pandas series
## And we're going to use the same index as our test dataset
```

```python
import pandas as pd
```

```python
preds = pd.Series(preds, index=test.index) ## Creating the series
```

```python
preds ### The predictions is now easy to see
```

```python
precision_score(test["Target"], preds)
```

```python
### Plotting our predictions, we will combine our actuals values with our predicted values
### Using the pandas concat, so we're concatinating our test target(which is our actual values) against our predicted values
### And then pass in the axis=1 which means treat each input as a column in our dataset
combined  = pd.concat([test["Target"], preds], axis=1)
```

### Building a BackTesting system

```python
### Building a backtracking system, this will basically just wrap up everything we just did into one function
### Fitting the model using the training predictors and the target, then combining our model into a series
### Then combining everything together then returning our combined dataframe
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
### Writing a backtest function which takes in our sp500 data, our ml model, our predictors,
### It also takes in a start value set to 2500 and step value.
### When you backtest, you want to a certain type of data to train your first model
### Every trading year has about 250 days, so the 2500 is like saying take 10 years of data,
### then train your first model using the 10 years of data
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    ### Function to loop across our data year by year and make predictions for all the years except the first 10 or so
    for i in range(start, data.shape[0], step) :
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions) ## Concatinating all our predictions together
```

```python
### We then backtest for our sp500 data with the models we created earlier and then predictors we created earlier
predictions = backtest(sp500, model, predictors)
```

```python
## Evaluating the error of our predictions
## Checking the predictions to check how we've predicted the days the market will go up or down
predictions["Predictions"].value_counts()
### value_counts, counts how many times each prediction was made
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
### is the 53% good or not, as a benchmark we can look the at the no. of days the market really went up
predictions["Target"].value_counts() / predictions.shape[0]
```

### Adding more predictors to our model

```python
### Creating a variety of rolling averages
horizons = [2,5,60,250,1000]
new_predictors = [] ### Will hold some of the new columns that we're goin to create

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"] ## Adding it to our sp500 dataset
    
    trend_column = f"Trend_{horizon}" ## trend is the number of days in the past x days that the stock price actually went up
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

### Improving our model

```python
### updating our model, increasing our no of estimators to 200 and reducing our min_samples_split to 50
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
### Slightly rewriting our predict function
### More control on defining what becomes a 1 and what becomes a zero using the predict_proba
### It will return the probability that the row will be a 0 or 1 
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors]) [:,1]
    preds[preds >= .6] = 1 ## setting our custom threshold
    preds[preds < .6] = 0 ## setting it to 60%, 
    ### model has to be more confident that the price will go up in order to show that it will go up
    ### this will reduce our total no. of trading days but it will increase the chance the price will actually go up in those days
    ### we don't want to make a ton of trades, we want to know that if we make a trade the price will go up
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
## Rerunning our backtest
predictions = backtest(sp500, model, new_predictors)
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
